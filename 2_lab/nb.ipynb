- розглянемо послідовні типи даних
> `list` - через те що він mutable його можна змінювати коли захочеш;

> `tuple` - а він є immutable і змінювати його не можна;

````python
l = [1, "2", 3]
ll = list([1, "2", 3])
print("Маємо список, який складається з елементів:","\nl=",l,"\nll=",ll)

l.append(["New", "List"])
print("Тепер список буде:", l)
l.insert(4, "Це буде в кінці")
print("Після вставки елементу на позицію перед 4 будемо мати:", l)

d = list(l.pop(-2)).append(-1,"тепер це останній елемент")
print("Елементи масиву зараз:", l)
print("Те що ми видалили з списку та модифікували буде:", d)
````

> Спитались в ChatGPT чому так, ось пояснення:

Програма на Python, яку ви подали, виведе `None`, і я поясню, чому це сталося.

Ось ваш код:

```python
l = [1, "2", 3]
d = list(l.pop(-2)).append("тепер це останній елемент")
```

Давайте подивимося на кожен рядок окремо:

1. `l = [1, "2", 3]`: В цьому рядку ви створюєте список `l` з трьома елементами: цілочисельним числом 1, рядком "2" і цілочисельним числом 3.

2. `l.pop(-2)`: Тут ви використовуєте метод `pop()`, щоб видалити другий елемент зі списку `l`. Однак `l.pop(-2)` повертає видалений елемент, тобто рядок "2".

3. `list(l.pop(-2))`: Ви намагаєтеся перетворити рядок "2" (повернутий результатом попереднього виразу) в список. Це призводить до створення списку, який містить окремі символи з рядка, тобто `['2']`.

4. `.append("тепер це останній елемент")`: Ви намагаєтеся викликати метод `append()` на створеному списку в пункті 3. Однак метод `append()` модифікує список на місці і не повертає нового списку. Таким чином, вираз `list(l.pop(-2)).append("тепер це останній елемент")` додає рядок "тепер це останній елемент" до списку `['2']`, але результат цього виразу - це `None`, тобто нічого.

Отже, після виконання цієї програми, змінна `d` буде містити `None`, а список `l` буде містити `[1, 3]`.


````python
t = (1, "2", 3)
tt = tuple([1, "2", 3])
print("Створений tuple буде мати такий вигляд завжди:", t)

print("До tuple можна застосовуватилише пошукові методи, наприклад знайти індекс елмента 3: ", t.index(3))
````

- знайомство з словниками

````python
a = "Це значення змінної а"
l = [1, "List", 2]
k = "k_k"
d = {"key": "value", 1: 2, "Це є ключ": "Це його значення", "a": a, "l": l, k: k*5}
print("Утворений словник буде мати вигляд:\n", d)

print("У списках ми доступаємось по індексах, беремо індекс 1:",l[1],
      "\nВ словниках для доступу до елемента використовуємо його ключ, беремо ключ 'Це є ключ':",d["Це є ключ"])

k2 = 'k_k'
print("Навіть якщо використати нову змінну, але в якій буде записано правильний ключ, ми зможемо доступитись до значення з словника.",
      "\nЗначення буде:",d[k2])

d.pop(k2)
print("Після видалення буде:",d)
print("Можемо переглянути всі наявні в словнику ключі:", d.keys())
print("Можемо переглянути всі наявні в словнику значення:", d.values())

dd = {"Прізвище": ["Пупкін", "Персона_2"], "імя": ["Іван", "Ілон"]}
print("Приклад складних типів даних можу бути таким", dd,
      "\nДоступатись до нього ми можемо ось так:", dd["Прізвище"][0])
````
